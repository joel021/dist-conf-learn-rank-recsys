```python
import torch
import matplotlib.pyplot as plt
```

```python
c = torch.arange(0, 1, 0.01)
delta_t = 0.125
standard_dist = torch.distributions.normal.Normal(0,1)

sigma = torch.sqrt(delta_t / (2 * standard_dist.icdf((1 + c) / 2)))
by_exp = torch.exp(-c-2)
by_log = delta_t/2-torch.log(c)
by_log2 = -torch.log(c)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(c, sigma, color='blue', linestyle='-', linewidth=2)
plt.plot(c, by_exp, color='green', linestyle='-', linewidth=2)
plt.plot(c, by_log, color='red', linestyle='-', linewidth=2)
plt.plot(c, by_log2, color='yellow', linestyle='-', linewidth=2)
plt.axhline(delta_t, color='red')
# Add labels and title
plt.title("title")
plt.xlabel("confidence")
plt.ylabel("sigma")

# Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```
